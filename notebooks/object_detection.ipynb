# Detecção de objetos

```python
import numpy as np 
import cv2 
from matplotlib import pyplot as plt
```

```python
def image_detector(xml:str, imagem_path:str, scaleFactor:float=1.2, minNeighbors:int=5):

    detector = cv2.CascadeClassifier(xml)
    imagem = cv2.imread(imagem_path)
    imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_RGB2GRAY)

    deteccoes = detector.detectMultiScale(imagem_cinza, scaleFactor, minNeighbors)

    for (x, y, w, h) in deteccoes:
        cv2.rectangle(imagem, (x, y), (x + w, y + h), (0, 255, 0), 2)
    
    return plt.imshow(imagem);
```
