```python
# libs
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

# Dados

```python
path = '/home/eric/projetos pessoais/computer-vision/Visão Computacional Guia Completo/Images/people1.jpg'
```

# Carregar imagem

```python
imagem = cv2.imread(path)
```

```python
imagem.shape
```

```python
obj_img = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(obj_img)
plt.show()
```

```python
# mudar tamanho para detecção correta
obj_img = cv2.resize(obj_img, (800,600))
```

```python
obj_img.shape
```

```python
plt.imshow(obj_img);
```

```python
# converter para escala de cinza
imagem_cinza = cv2.cvtColor(obj_img, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(imagem_cinza, cmap = 'gray', interpolation = 'bicubic')
plt.show()
```

# Detector facial

```python
xml_path ='/home/eric/projetos pessoais/computer-vision/Visão Computacional Guia Completo/Cascades/haarcascade_frontalface_default.xml'

detector_facial = cv2.CascadeClassifier(xml_path)
```

```python
deteccoes = detector_facial.detectMultiScale(imagem_cinza, scaleFactor=1.09)
```

```python
deteccoes
```

```python
for x,y,w,h in deteccoes:
    cv2.rectangle(obj_img, (x,y), (x + w,y + h), (255, 255, 0), 5)
plt.imshow(obj_img);
```

```python
# ler segunda imagem
imagem2 = cv2.imread('/home/eric/projetos pessoais/computer-vision/Visão Computacional Guia Completo/Images/people2.jpg')
imagem2 = cv2.cvtColor(imagem2, cv2.COLOR_BGR2RGB)
# converter para escala de cinza
imagem_cinza_2 = cv2.cvtColor(imagem2, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(imagem_cinza_2, cmap = 'gray', interpolation = 'bicubic')
plt.show()
```

```python
deteccoes = detector_facial.detectMultiScale(imagem_cinza_2, scaleFactor=1.2, minNeighbors=3, minSize=(30,30))
```

```python
len(deteccoes)
```

```python
for x,y,w,h in deteccoes:
    cv2.rectangle(imagem2, (x,y), (x + w,y + h), (255, 255, 0), 2)
plt.imshow(imagem2);
```
